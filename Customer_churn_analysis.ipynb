```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fact=pd.read_csv('/content/Fact_CustomerChurn.csv')
date=pd.read_csv('/content/Dim_Date.csv')
fact
```

```python
date
```

```python
df=fact.merge(date, on='DateID')
```

```python
df
```

```python
df['FullDate']=pd.to_datetime(df['FullDate'])
```

```python
churn=df[df['Churned']=='Yes']
churn
```

```python
monthly_churn=(
    churn.sort_values('FullDate').assign(Month=lambda x: x['FullDate'].dt.strftime('%b %y'))
    .groupby('Month',sort=False).size().reset_index(name='churncount')
)
```

```python
monthly_churn
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(12,6))
plt.fill_between(
    monthly_churn['Month'],
    monthly_churn['churncount'],
    color='lightgreen',
    alpha=0.5
)
plt.plot(
    monthly_churn['Month'],
    monthly_churn['churncount'],
    color='green',
    linewidth=2
)
# data labels
for i,row in monthly_churn.iterrows():
  plt.text(
  i,
  row['churncount']+1,
  str(row['churncount']),
  ha='center',
  fontsize=9)
  plt.title('Monthly Customer Churn Trend')
  plt.xlabel('Month')
  plt.ylabel('Number of churned customers')
  plt.xticks(
     ticks=range(len(monthly_churn)),
     labels=monthly_churn['Month'],
     rotation=45
  )
  plt.grid(True,linestyle='--',alpha=0.6)
  plt.tight_layout()
```

Daily Customer churn

```python
df['DayOfWeek'] = df['DayOfWeek'].str.replace('\t', '').str.strip()
```

```python
daily=df[df['Churned']=='Yes'].groupby('DayOfWeek').size().reindex([
    'Monday',
    'Tuesday',
    'Wednesday',
    'Thursday',
    'Friday',
    'Saturday',
    'Sunday'
]).fillna(0)
```

```python
daily
```

```python
plt.figure(figsize=(8,5))
plt.bar(daily.index,daily.values)
plt.title('Daily Customer churn')
plt.xlabel('Day')
plt.ylabel('number of churned customers')
plt.xticks(rotation=45)
plt.show()
```

```python
# category wise net revenue
import pandas as pd
import plotly.express as px
fact=pd.read_csv('/content/Fact_CustomerChurn.csv')
product=pd.read_csv('/content/Dim_Customer.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
fact=pd.read_csv('/content/Fact_CustomerChurn.csv')
product=pd.read_csv('/content/Dim_Product.csv')
```

```python
df=fact.merge(product, on='ProductID')
df
```

```python
revenue=(
    df.groupby('ProductName')['NetRevenue'].sum().sort_values()/1000000)
revenue
```

```python
plt.figure(figsize=(10,6))
bars=plt.barh(
    revenue.index,
    revenue.values,
    color='skyblue'
)
for x in bars:
  plt.text(
      x.get_width(),
      x.get_y()+x.get_height()/2,
      f'{x.get_width():.1}M',
      va='center'
  )
  plt.title('Total Revenue by Product Revenue')
  plt.xlabel('Revenue (Million)')
  plt.ylabel('Product Category')
```
